### Jag skapar ett mindre dataset för att bilda ett model

```python
df_small = df[['Age','Debt','YearsEmployed','Employed','CreditScore','Income','Approved','Gender_a','Gender_b']]
df_small 
```

```python
import seaborn as sns
sns.heatmap(df_small.corr(),annot=True, fmt='.2f')
```

```python
x = np.array(df_small.drop(['Approved'],axis=1))
y = np.array(df_small['Approved'])
```

```python
x
```

### Split the dataset into training and test dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
#now model is ready after training the data
```

```python
linear_model.score(x_test,y_test)
```
